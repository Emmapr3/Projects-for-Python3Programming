**1.- Below are a set of scores that students have received in the past semester. 
This code determines how many are 'min_score' or above and assign that result to the value 'a_scores'.**

```python
scores = "67 80 90 78 93 20 79 89 96 97 92 88 79 68 58 90 98 100 79 74 83 88 80 86 85 70 90 100"
a_scores = 0
min_score = 90
for score in scores.split(" "): #scores.split(" ") creates a list of scores like string type 
    if int(score) >= min_score:
        a_scores += 1  #adds 1 if the score is grather or equal than min_score
print(a_scores) #prints the total number of scores gather or equal than min_score
```

**2.-This code uses the string stored in the 'org' variable and creates an acronym which is assigned to the variable 'acro'. Only the first letter of each word should be used, each letter in the acronym should be a capital letter and there should be nothing to separate the letters of the acronym. Words that should not be included in the acronym are stored in the list 'stopwords'.**
For example, if org was assigned the string “hello to world” then the resulting acronym should be “HW”.


```python
def make_acro(str1):
  '''Function to generate a acronym of 'str1' only taking the fisrt letter of each word (spotwords list not included)
    as capital letter and without separation beetween each letter'''
  stopwords = ['to', 'a', 'for', 'by', 'an', 'am', 'the', 'so', 'it', 'and','its']
  acro = '' 
  for word in str1.split(): #takes each word in the list with the words of the string 'str1'
    if word.lower() not in stopwords: #revises if the word in lower case is not in spotwords list
      acro += word[0].upper() #adds the first letter of the word as capital case
  return acro

org = 'The organization for health, safety, and education'
print(org,'--',make_acro(org))
org2 = 'United Nations Educational, Scientific and Cultural Organization'
print(org2,'--',make_acro(org2))
org3 = 'International Business Machines'
print(org3,'--',make_acro(org3))
```

**3.- This code take the string stored in 'sent' variable and creates an acronym which is assigned to the variable 'acro'. The first two letters of each word should be used, each letter in the acronym should be a capital letter, and each element of the acronym should be separated by a “. ” (dot and space). Words that should not be included in the acronym are stored in the list stopwords.** For example, if sent was assigned the string “height and ewok wonder” then the resulting acronym should be “HE. EW. WO”.
# Nueva sección

```python
stopwords = ['to', 'a', 'for', 'by', 'an', 'am', 'the', 'so', 'it', 'and']
sent = "The water earth and air are vital"
acro = '' 
sent_lst = [word for word in sent.split() if word.lower() not in stopwords] #makes a list of words in 'sent' not included in stopwords list
for word in sent_lst:
    acro += word[:2].upper()
    if word != sent_lst[-1]: #to avoid add '. ' to the last word
            acro += '. '            
print(acro)
```

**4.- A palindrome is a phrase that, if reversed, would read the exact same. This code checks if a phrase 'p_phrase' is a palindrome by reversing it and then checking if the reversed version is equal to the original with a function.**

```python
def IsPalindrome(p_phrase):
  '''This function returns if the phrase 'p_phrase' is a palindrome, True or False'''
  r_phrase = p_phrase[::-1] #reverts the direction of the phrase with all letters
  p_phrase_joined = ''.join(p_phrase.split()) #joins the p_phrase without spaces as string 
  r_phrase_joined = ''.join(r_phrase.split()) #joins the r_phrase without spaces as string
  test = p_phrase_joined.lower()==r_phrase_joined.lower() #lower() avoids the differences by capital cases 
  return test

p_phrase = "was it a car or a cat I saw"  #a palindrome on English
print(IsPalindrome(p_phrase))
print(IsPalindrome('Hi, how arre you?')) #evidently this is not a palindrome
print(IsPalindrome('Anita lava la tina')) #very popular palindrome on Spanish
```

**5.- Provided is a list of data about a store’s inventory where each item in the list represents the name of an item, how much is in stock, and how much it costs. This code prints out each item in the list with the same formatting, using the .format method (not string concatenation).** For example, the first print statment should read The store has 12 shoes, each for 29.99 USD.


```python
inventory = ["shoes, 12, 29.99", "shirts, 20, 9.99", "sweatpants, 25, 15.00", "scarves, 13, 7.75"]
for elem in inventory:
    info_article = tuple(elem.split(',')) #split every element article intro a tuple
    print('The store has{1} {0}, each for{2} USD.'.format(*info_article)) #prints the info of each article in the same format of the example above
```
