## Player Classes

* The **Class WOFPlayer** represents a Wheel of Fortune Player. Every instance of *WOFPlayer* has three instance variables: *name*, *prizeMoney* and *prizes*. Aditionally, it should have the following methods: *addMoney(amt)*, *gobankrup()* and *addPrizes(prize)*.

* The **Class WOFHumanPlayer** inherits the previous class and represent a human player that plays through the prompt. Also, it adds a method for requesting a valid move.

* The **Class WOFComputerPlayer** is the most interesting. Also, it inherits the parent class WOFPlayer and adds one instance variable to the player: the *difficulty*. Computer player with a higher difficulty generally plays "better". This difficulty is between 1 (easier opponent) and 10 (harder opponent). To decide if the Computer player makes a "good" or "bad" move, a semi-randomly process is used. A "bad" move refers to a random decision on possible letters. In a "good" move, a letter according to their overall frequency in the English language is chosen. To do this last, a class variable 'SORTED_FREQUENCIES' might be included, where 'SORTED_FREQUENCIES = ZQXJKVBPYGFWMUCLDRHSNIOATE' is the list of English characters sorted from the least frequent (Z) to the most frequent (E). In this class, *smartCoinFlip()* method is used to decide if the Computer player makes a bad or good move according to its difficulty level: choosing a random number between 1 and 10, if this number is lesser or equal than its difficulty then the player makes a good move, otherwise makes a bad move. Moreover, the *getPossibleLetters(guessed)* method returns a list of letters that can be guessed, considering the letters already guessed and whether the player has enough money to guess vowels (*moneyPrize*>=*VOWEL_COST*). For the player to make a good or bad move, this list of possible letters to guess must be considered.

```python
# CLASS for a general WOF's Player (petern class)
class WOFPlayer:
    #constructor, name is only the parameter to the WOFPlayer object
    def __init__(self, name):
        self.name = name #name of player
        self.prizeMoney  = 0 #amount of prize money for this player
        self.prizes = [] #the prizes this plater has won fo far
    #add the amount amt to self.prizeMoney    
    def addMoney(self, amt):
        self.prizeMoney += amt
    #set self.prizeMoney to 0    
    def goBankrupt(self):
        self.prizeMoney = 0
    #append prize to self.prizes list    
    def addPrize(self, prize):
        self.prizes.append(prize)
    # print: name(prizeMoney) 
    def __str__(self):
        return '{} (${})'.format(self.name, self.prizeMoney)
    
# Class for a WOF's human player, inherits the previous class and only add a method for requesting a player move
class WOFHumanPlayer(WOFPlayer):
    #method to get valid move from the Human Player
    def getMove(self, category, obscuredPhrase, guessed):
        prompt = """
        {} has ${}
        
        Category: {}
        Phrase:   {}
        Guessed:  {}
        
        Guess a letter, phrase, or type 'exit' or 'pass':""".format(self.name, self.prizeMoney, category, obscuredPhrase, ', '.join(sorted(guessed)))
        return input(prompt)
    
# Class for a WOF's computer player, inherits the parent class and add the variable instance 'difficulty' to assing how well computer player will play.
class WOFComputerPlayer(WOFPlayer):
    #list of English characters sorted from least frequent ('Z') to most frequent ('E').
    SORTED_FREQUENCIES = 'ZQXJKVBPYGFWMUCLDRHSNIOATE'
    
    #constructor
    def __init__(self, name, difficulty):
        WOFPlayer.__init__(self,name)
        self.difficulty = difficulty #higher difficulty generally play 'better'
    #method to decide semi-randomly wheter to make a 'good' or 'bad' move
    def smartCoinFlip(self):
        return random.randint(1,10) <= self.difficulty
    #method to retunr a list of letter that can be guessed
    def getPossibleLetters(self, guessed):
        possibleLetters = [ch for ch in LETTERS if ch not in guessed] #LETTERS\guessed
        if self.prizeMoney < VOWEL_COST: #player can't pay the VOWEL
            possibleLetters = [ch for ch in possibleLetters if ch not in VOWELS] #possibleLetters\VOWELS
        return possibleLetters
    #method to get a valid move from the Computer Player
    def getMove(self,category, obscuredPhrase, guessed):
        possibleGuessed = self.getPossibleLetters(guessed)
        #possibleGuessed are all vowels and prizeMoney < VOWEL_COST
        #getPossibleLetters() return empty list
        if len(possibleGuessed) == 0:
            return 'pass'
        #good (true) or bad (false) move
        if self.smartCoinFlip():
            #return the most frequent (highest index in SORTED_FREQUENCIES) possible character
            sorted_freq_possibleGuessed = sorted(possibleGuessed, key = lambda x: self.SORTED_FREQUENCIES.index(x))
            return sorted_freq_possibleGuessed[-1]
        else:
            #return a random char from possibleGuessed
            return random.choice(possibleGuessed)
```

## Game implementatio in code

A set of functions are defined to help in the game code implementation:
* getNumberBetween: to ask the user for the number of Human players and Compuer Players (only accept valid numbers).
* spinWheel: to choice randomly a prize from the WOF. The json file 'wheel.json' has all possible prizes.
* getRandomCategoryAndPhrase: to obtain randomly a category and a phrase from this category to guess.
* obscurePhrase: to return a obscured version of the phrase with the guessed letters (see example in the code)
* showBoard: to show the current state of the game

The file **'wheel.json'** include the possible prizes/squares in the wheel of fortune.

The file**'phrases.json'** has the caterogies and their phrases for guess in the game.

```python
import json
import random
import time

LETTERS = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
VOWELS  = 'AEIOU'
VOWEL_COST  = 250

# Repeatedly asks the user for a number between min & max (inclusive)
def getNumberBetween(prompt, min, max):
    userinp = input(prompt) # ask the first time (prompt text)

    while True:
        try:
            n = int(userinp) # try casting to an integer
            if n < min:
                errmessage = 'Must be at least {}'.format(min)
            elif n > max:
                errmessage = 'Must be at most {}'.format(max)
            else:
                return n
        except ValueError: # The user didn't enter a number
            errmessage = '{} is not a number.'.format(userinp)

        # If we haven't gotten a number yet, add the error message
        # and ask again
        userinp = input('{}\n{}'.format(errmessage, prompt))

# Spins the wheel of fortune wheel to give a random prize
# Examples:
#    { "type": "cash", "text": "$950", "value": 950, "prize": "A trip to Ann Arbor!" },
#    { "type": "bankrupt", "text": "Bankrupt", "prize": false },
#    { "type": "loseturn", "text": "Lose a turn", "prize": false }
def spinWheel():
    with open("wheel.json", 'r') as f: #open file with the possible prizes in the wheel
        wheel = json.loads(f.read())
        return random.choice(wheel)

# Returns a category & phrase (as a tuple) to guess
# Example:
#     ("Artist & Song", "Whitney Houston's I Will Always Love You")
def getRandomCategoryAndPhrase():
    with open("phrases.json", 'r') as f: #open file with the categories and phrases
        phrases = json.loads(f.read())

        category = random.choice(list(phrases.keys()))
        phrase   = random.choice(phrases[category])
        return (category, phrase.upper())

# Given a phrase and a list of guessed letters, returns an obscured version
# Example:
#     guessed: ['L', 'B', 'E', 'R', 'N', 'P', 'K', 'X', 'Z']
#     phrase:  "GLACIER NATIONAL PARK"
#     returns> "_L___ER N____N_L P_RK"
def obscurePhrase(phrase, guessed):
    rv = ''
    for s in phrase:
        if (s in LETTERS) and (s not in guessed):
            rv = rv+'_'
        else:
            rv = rv+s
    return rv

# Returns a string representing the current state of the game
def showBoard(category, obscuredPhrase, guessed):
    return """
Category: {}
Phrase:   {}
Guessed:  {}""".format(category, obscuredPhrase, ', '.join(sorted(guessed)))
```

```python
# GAME LOGIC CODE
print('='*26)
print('WHEEL OF FORTUNE IN PYTHON')
print('='*26)
print('')

num_human = getNumberBetween('How many human players? ', 0, 10) #ask number of human players

# Create the human player instances. Assigned by the players' name
human_players = [WOFHumanPlayer(input('Enter the name for human player #{}'.format(i+1))) for i in range(num_human)]

num_computer = getNumberBetween('How many computer players? ', 0, 10) #ask the number of computer players

# If there are computer players, ask how difficult they should be
if num_computer >= 1:
    difficulty = getNumberBetween('What difficulty for the computers? (1-10)', 1, 10)

# Create the computer player instances, using the previous difficulty list
computer_players = [WOFComputerPlayer('Computer {}'.format(i+1), difficulty) for i in range(num_computer)]

players = human_players + computer_players #total number of players

# No players, no game :(
if len(players) == 0:
    print('We need players to play!')
    raise Exception('Not enough players')

# category and phrase are strings.
category, phrase = getRandomCategoryAndPhrase()
# guessed is a list of the letters that have been guessed
guessed = []

# playerIndex keeps track of the index (0 to len(players)-1) of the player whose turn it is
playerIndex = 0

# will be set to the player instance when/if someone wins
winner = False

def requestPlayerMove(player, category, guessed):
    while True: # we're going to keep asking the player for a move until they give a valid one
        time.sleep(0.1) # added so that any feedback is printed out before the next prompt

        move = player.getMove(category, obscurePhrase(phrase, guessed), guessed) #getMove for each class
        move = move.upper() # convert whatever the player entered to UPPERCASE
        if move == 'EXIT' or move == 'PASS':
            return move
        elif len(move) == 1: # they guessed a character
            if move not in LETTERS: # the user entered an invalid letter (such as @, #, or $)
                print('Guesses should be letters. Try again.')
                continue
            elif move in guessed: # this letter has already been guessed
                print('{} has already been guessed. Try again.'.format(move))
                continue
            elif move in VOWELS and player.prizeMoney < VOWEL_COST: # if it's a vowel, we need to be sure the player has enough
                    print('Need ${} to guess a vowel. Try again.'.format(VOWEL_COST))
                    continue
            else:
                return move
        else: # they guessed the phrase
            return move


while True:
    player = players[playerIndex]
    wheelPrize = spinWheel()

    print('')
    print('-'*15)
    print(showBoard(category, obscurePhrase(phrase, guessed), guessed))
    print('')
    print('{} spins...'.format(player.name))
    time.sleep(2) # pause for dramatic effect!
    print('{}!'.format(wheelPrize['text']))
    time.sleep(1) # pause again for more dramatic effect!

    if wheelPrize['type'] == 'bankrupt':
        player.goBankrupt()
    elif wheelPrize['type'] == 'loseturn':
        pass # do nothing; just move on to the next player
    elif wheelPrize['type'] == 'cash':
        move = requestPlayerMove(player, category, guessed)
        if move == 'EXIT': # leave the game
            print('Until next time!')
            break
        elif move == 'PASS': # will just move on to next player
            print('{} passes'.format(player.name))
        elif len(move) == 1: # they guessed a letter
            guessed.append(move)

            print('{} guesses "{}"'.format(player.name, move))

            if move in VOWELS:
                player.prizeMoney -= VOWEL_COST

            count = phrase.count(move) # returns an integer with how many times this letter appears
            if count > 0:
                if count == 1:
                    print("There is one {}".format(move))
                else:
                    print("There are {} {}'s".format(count, move))

                # Give them the money and the prizes
                player.addMoney(count * wheelPrize['value'])
                if wheelPrize['prize']:
                    player.addPrize(wheelPrize['prize'])

                # all of the letters have been guessed
                if obscurePhrase(phrase, guessed) == phrase:
                    winner = player
                    break

                continue # this player gets to go again

            elif count == 0:
                print("There is no {}".format(move))
        else: # they guessed the whole phrase
            if move == phrase: # they guessed the full phrase correctly
                winner = player

                # Give them the money and the prizes
                player.addMoney(wheelPrize['value'])
                if wheelPrize['prize']:
                    player.addPrize(wheelPrize['prize'])

                break
            else:
                print('{} was not the phrase'.format(move))

    # Move on to the next player (or go back to player[0] if we reached the end)
    playerIndex = (playerIndex + 1) % len(players)

if winner:
    # In your head, you should hear this as being announced by a game show host
    print('{} wins! The phrase was {}'.format(winner.name, phrase))
    print('{} won ${}'.format(winner.name, winner.prizeMoney))
    if len(winner.prizes) > 0:
        print('{} also won:'.format(winner.name))
        for prize in winner.prizes:
            print('    - {}'.format(prize))
else:
    print('Nobody won. The phrase was {}'.format(phrase))
```
